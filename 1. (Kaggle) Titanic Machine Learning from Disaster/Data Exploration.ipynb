```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# initialize logger
import logging
logging.basicConfig(level = logging.INFO,
                    format = '%(asctime)s - %(levelname)s - %(message)s')
```

```python
# Read Dataset from file
dataset = pd.read_csv('dataset/train.csv')
logging.info('Dataset Shape : ' + str(dataset.shape) )
```

```python
dataset.head(5)
```

## Analysis of our dataset

```python
number_of_unique_names = len(dataset['Name'].value_counts())
logging.info('Number of Unique names : ' + str(number_of_unique_names) )
```

```python
dataset['Survived'].value_counts().sort_index().plot.bar()
plt.title('Count of survivors and Non-survivors')
plt.xlabel('0 - Not Survived , 1- Survived')
plt.ylabel('Number of people')
```

```python
dataset['Pclass'].value_counts().sort_index().plot.bar()
plt.title('Count of passengers in different class')
plt.xlabel('Class')
plt.ylabel('Number of people')
```

```python
dataset['Sex'].value_counts().sort_index().plot.bar()
plt.xlabel('Sex')
plt.ylabel('Number of people')
```

```python
dataset['Age'].value_counts().sort_index().plot.area()
plt.xlabel('Age (in years)')
plt.ylabel('Number of people')
```

```python
dataset['SibSp'].value_counts().sort_index().plot.bar()
plt.xlabel('# of siblings / spouses aboard the Titanic')
plt.ylabel('Number of people')
```

```python
dataset['Parch'].value_counts().sort_index().plot.bar()
plt.xlabel('# of parents / children aboard the Titanic')
plt.ylabel('Number of people')
```

```python
dataset['Fare'].value_counts().sort_index().plot.area()
plt.xlabel('Passenger Fare')
plt.ylabel('Number of people')
```

```python
dataset['Embarked'].value_counts().sort_index().plot.bar()
plt.xlabel('Passenger Fare')
plt.ylabel('Number of people')
```

```python
number_of_unique_names = len(dataset['Cabin'].value_counts())
logging.info('Number of Unique names : ' + str(number_of_unique_names) )
```

## Relation of features with survival 

```python
dataset.plot.scatter(x='Age', y='Survived')
plt.xlabel('Age')
plt.ylabel('0 - Not Survived , 1- Survived')
```

```python
dataset.plot.scatter(x='Pclass', y='Survived')
plt.xlabel('Pclass')
plt.ylabel('0 - Not Survived , 1- Survived')
```
